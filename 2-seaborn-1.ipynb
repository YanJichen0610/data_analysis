```python
import seaborn as sns
import numpy as np
import matplotlib as mpl
import matplotlib.pyplot as plt
import scipy.stats as stats
import pandas as pd
%matplotlib inline
```

直方图 (Histogram)

描述: 直方图用于显示数据的分布情况，通过将数据划分为若干区间（bins），然后计算每个区间内的数据点数量。
适用场景: 适合查看数值型数据的频率分布、识别数据的中心趋势、分散程度和偏斜度。

```python
# 生成数据
x = np.random.gamma(6, size=400)

# 创建画布和子图
plt.figure(figsize=(5, 2))

# 第一个子图：默认bins数量的直方图
plt.subplot(1, 2, 1) # 1行2列的第一个图
sns.histplot(x, kde=False, color="skyblue", stat="density")
plt.title('Default Bins')

# 第二个子图：指定bins数量的直方图
plt.subplot(1, 2, 2) # 1行2列的第二个图
sns.histplot(x, kde=False, bins=40, color="olive", stat="density")
plt.title('40 Bins')

# 调整布局确保不被裁切
plt.tight_layout()

# 显示图形
plt.show()
```

```python
# 设置均值向量 [x的均值, y的均值]
mean = [0, 5]

# 设置协方差矩阵：[[x的方差, xy协方差], [xy协方差, y的方差]]
cov_matrix = [
    [1.0, 0.8],
    [0.8, 2.0]
]

# 生成100个样本
num_samples = 100
xy = np.random.multivariate_normal(mean=mean, cov=cov_matrix, size=num_samples)

# 转换为 Pandas DataFrame
df = pd.DataFrame(xy, columns=['x', 'y'])

# 输出前5行数据
print("生成的数据前5行：")
print(df.head())

# 可选：输出统计信息
print("\n基本统计信息：")
print(df.describe())
```
