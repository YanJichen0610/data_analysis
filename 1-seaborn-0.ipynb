```python
import seaborn as sns
import numpy as np
import matplotlib as mpl
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
#  sinplot  flip 1
# flip  1  -1 
def sinplot(flip=1):
    #  numpy  linspace  0  14  100 
    #  x 
    x = np.linspace(0, 14, 100)
    
    #  i  1  6 6 
    for i in range(1, 7):
        #  i  matplotlib  plot 
        # np.sin(x + i * 0.5) i*0.5  x 
        # (7 - i) flip 
        plt.plot(x, np.sin(x + i * 0.5) * (7 - i) * flip)


sinplot()
```
