```python
import seaborn as sns
import numpy as np
import matplotlib as mpl
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
# 定义一个名为 sinplot 的函数，并接受一个可选参数 flip，默认值为 1。
# flip 参数可以用来控制图形的上下翻转（乘以 1 不变，乘以 -1 则上下翻转）。
def sinplot(flip=1):
    # 使用 numpy 的 linspace 函数创建一个从 0 到 14 的等差数列，共 100 个点。
    # 这些点将作为 x 轴的数据点。
    x = np.linspace(0, 14, 100)
    
    # 循环 i 从 1 到 6（包括），即一共循环 6 次，每次绘制一条不同的正弦曲线。
    for i in range(1, 7):
        # 对于每一个 i 值，使用 matplotlib 的 plot 函数绘制一条正弦曲线。
        # np.sin(x + i * 0.5) 计算正弦值，i*0.5 使得每条曲线在 x 轴上偏移，
        # (7 - i) 使得每条曲线有不同的振幅，flip 参数用于控制图形的翻转。
        plt.plot(x, np.sin(x + i * 0.5) * (7 - i) * flip)


sinplot()
```

```python
# seabeon的默认颜色
# sns.palplot 是 Seaborn 库中的一个函数，专门用于可视化颜色调色板
current_palette = sns.color_palette()
sns.palplot(current_palette)
```

```python
# 现在我想要设定默认颜色的种类，可以使用以下方法
# 假如我要做一个八分类的图片 我就传入一个8
sns.palplot(sns.color_palette("hls", 6))
```

```python
# 第一张图（定义了4种颜色）
sns.boxplot(data=data, palette=sns.color_palette("hls", 4))
plt.title("Palette with 4 colors")
plt.show()

# 清除当前图像
plt.clf()

# 第二张图（定义了8种颜色）
sns.boxplot(data=data, palette=sns.color_palette("hls", 8))
plt.title("Palette with 8 colors")
plt.show()
```
